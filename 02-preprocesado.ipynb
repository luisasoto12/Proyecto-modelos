#**Número de valores nulos por columna e imputación**

```python
k = d.isna().sum()
k[k!=0]
```

```python
d_limpio = d.copy()
```

```python
#Para las variables numéricas faltantes rellenamos con el promedio
imputer_num = SimpleImputer(strategy='mean')
num = d_limpio.select_dtypes(include=['int64', 'float64']).columns
d_limpio[num] = imputer_num.fit_transform(d_limpio[num])

#Para las variables categórica rellenamos con el valor más frecuente
imputer_cte = SimpleImputer(strategy='most_frequent')
cte = d_limpio.select_dtypes(include=['object']).columns
d_limpio[cte] = imputer_cte.fit_transform(d_limpio[cte])
```

```python
# Número de columnas con valores nulos
num_columnas_nulas = (d_limpio.isna().sum() > 0).sum()
print(f"Número de columnas con valores nulos: {num_columnas_nulas}")
```

```python
k = d_limpio.isna().sum()
k
```

#**Eliminación de columnas irrelevantes**

```python
d_limpio = d_limpio.copy()
columns= ['F_TIENELAVADORA','F_TIENEAUTOMOVIL','E_PRIVADO_LIBERTAD']
d_limpio = d_limpio.drop(columns = [c for c in columns if c in d_limpio.columns])
d_limpio.head()
```

#**Normalización**

```python
d_limpio = d_limpio.copy()
for column in ['F_EDUCACIONMADRE', 'F_EDUCACIONPADRE', 'E_VALORMATRICULAUNIVERSIDAD']:
    d_limpio[column] = d_limpio[column].replace(['No sabe', 'No Aplica'], 'No aplica')

d_limpio[['F_EDUCACIONMADRE', 'F_EDUCACIONPADRE', 'E_VALORMATRICULAUNIVERSIDAD']].head()
```

#**Conversión de columnas catégoricas en one hot**


```python
d_limpio = d_limpio.copy()
colms = ['E_PRGM_ACADEMICO','E_PRGM_DEPARTAMENTO','F_EDUCACIONMADRE','F_EDUCACIONPADRE']
d_limpio = pd.get_dummies(d_limpio, columns=colms, drop_first=True)
d_limpio.head()
```

#**Conversión de variable objetivo a numérica**

```python
d_limpio['RENDIMIENTO_GLOBAL'].unique()
```

```python
d_limpio = d_limpio.copy()
academic_performance = {'bajo': 1, 'medio-bajo': 2.5, 'medio-alto': 3.5, 'alto': 5}
d_limpio['RENDIMIENTO_GLOBAL'] = d_limpio['RENDIMIENTO_GLOBAL'].map(academic_performance)
d_limpio['RENDIMIENTO_GLOBAL'].value_counts()
```
